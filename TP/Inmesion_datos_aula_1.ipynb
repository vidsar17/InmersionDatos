```python
# 2. Precio de m2 por barrio con el gráfico más adecuado para esta nueva variable

inmuebles['Valor_m2_Millon'] = inmuebles['Precio_Millon'] / inmuebles['Area']
inmuebles.head()
```

```python
# GROUP BY para agrupar

inmuebles.groupby('Barrio').mean()
```

```python
# tomamos los atributos de cada barrio y los sumamos y guardamos en la variables datos_barrio

datos_barrio = inmuebles.groupby('Barrio').sum()
datos_barrio
```

```python
# nuevo atributo Valor_m2_Barrio en el nuevo dataset datos_barrio

datos_barrio['Valor_m2_Barrio'] = datos_barrio['Precio_Millon'] / datos_barrio['Area']
datos_barrio
```

```python
# creamos un diccionario con los nuevos valores

dict(datos_barrio['Valor_m2_Barrio'])
```

```python
# llevamos el nuevo valor del m2 al original dataset

# creación de un diccionario
m2_barrio = dict(datos_barrio['Valor_m2_Barrio'])

# estamos duplicando una columna
inmuebles['Valor_m2_Barrio'] = inmuebles['Barrio']

# aquí estamos mapeando la columna duplicada con el diccionario
inmuebles['Valor_m2_Barrio'] = inmuebles['Valor_m2_Barrio'].map(m2_barrio)

# imprimimos el último valor
inmuebles.head(5)
```

```python
# top 10 de # top 10 de barrios en una variable

top_barrios = inmuebles['Barrio'].value_counts()[:10].index
top_barrios
```

```python
# convertimos el índice Barrio a columna y generamos un índice

datos_barrio.reset_index(inplace=True)
datos_barrio
```

```python
# aquí entrega el top 10 en un data frame

datos_barrio.query('Barrio in @top_barrios')
```

```python
# creamos un gráfico de barras

plt.figure(figsize = (10,6))
ax = sns.barplot(x = 'Barrio', y = 'Valor_m2_Barrio', data = datos_barrio.query('Barrio in @top_barrios'))

# tomamos los nombres de los barrios
ax.tick_params(axis = 'x', rotation = 45)
plt.show()
```

```python
# creamos un gráfico de cajas

plt.figure(figsize = (10,6))
ax = sns.boxplot(x = 'Barrio', y = 'Valor_m2_Millon', data = inmuebles.query('Barrio in @top_barrios'))

# tomamos los nombres de los barrios
ax.tick_params(axis = 'x', rotation = 45)
plt.show()
```
